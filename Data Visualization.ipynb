```python
movie = pd.read_csv("ml-latest/movies.csv")
movie.head(5)
```

```python
rating = pd.read_csv("ml-latest/ratings.csv")
rating.head(5)
```

```python
#Ways to convert a column to a standard format
movie["genres_cat"] =movie["genres"] .str.split("|")
movie.head()
```

```python
#Filter film by genre
filter1 = movie.apply(lambda row:"Children" in row["genres_cat"],axis=1)
Child_df = movie[filter1]
Child_df.head(5)
```

```python
#Count the number of movies according to their genre
genre_count = {}
for index, row in movie.iterrows():
    for genre in row["genres_cat"]:
        genre_count[genre] = genre_count.get(genre,0)+1
print(genre_count)
        
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(genre_count.values()),labels = genre_count.keys())
plt.show()
```

```python
unit_rate = rating["rating"].unique()
#scale of rating, maximum rating, minimum rating, and etc.
print(unit_rate)
#Number of users
len(rating["userId"].value_counts())
```

```python
df1 = rating.rating.value_counts()
x1 = np.linspace(0,5,num=11)
plt.figure(figsize = (8,6))
plt.hist(rating.rating,bins=10,color = "green")
plt.xticks(x1)
plt.title("Ratings")
plt.grid()
plt.show()
#Most of the user have given 4 rating to movies
```

```python
#merging both dataframe
merged_df = pd.merge(movie,rating,on="movieId")
merged_df.head(5)
```

```python
#Sorting movies containing movie id , title, and rating 
#Top 10 higly rated movies are shown 
mov_rating = merged_df[["movieId","rating","title"]]
mov_rating.groupby(["movieId","title"],as_index= False).mean().sort_values(by='rating',ascending = False).head(10)
```

```python
#Movie Brad Williams: Daddy Issues (2016) is in the top because 
#only two person voted 5 which is not convincing at all that 
#the movie deserve to be in top
len(merged_df[merged_df["movieId"] == 169338].index)
```

```python
# Now to solve this problem, first we need to
#count the number of rating a movie got and afterward will have
#to put one constraint on rating variable like rating>100 
#or rating>200 such that only those movies will be 
#considered which has got more than 100 or 200 ratings from users.
count_rate = mov_rating.groupby(["movieId","title"],as_index = False).count()
filter_count_rate = count_rate.apply(lambda x: x["rating"]>100,axis=1)
well_rated_df = count_rate[filter_count_rate]
well_rated_df.head(10)
```

```python
final_df = pd.merge(merged_df , well_rated_df, on =["movieId"])
mov_rating = final_df[["movieId","title_x","rating_x","rating_y"]]
X = mov_rating.groupby(["movieId","title_x"],as_index = False).mean().sort_values(by='rating_x',ascending = False)
X.head(10)
# Top 10 movies according to the rating and the number of people rate
```
